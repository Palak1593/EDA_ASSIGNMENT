```python
#Below are the libraries that are used to perform EDA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/data.csv')
df.head() #prints first five rows
```

```python
df.tail() #prints last five rows
```

```python
df.dtypes #retrieve the data types of each column in a DataFrame.
```

```python
pd.unique(df['MATS3s']) #pd.unique() in pandas is used to retrieve the unique values of a specific column in a DataFrame.
# pd.unique(df['MATS3s'])  is used to retrieve the unique values in the 'MATS3s' column of the DataFrame df.
```

```python
#Dropping irrelevant columns
df=df.drop('MATS3e',axis=1) #MAT3e is the column we are dropping
                          #axis= 1 or axis='index means you want to drop labels from the rows of the DataFrame.
                          #This operation removes the specified rows from the DataFrame, effectively reducing the number of rows.
df.head()
```

```python
#Dropping irrelevant rows
df=df.drop(3,axis=0) #we have dropped 3rd row
                     #axis=0 (default),means to drop labels from the columns of the DataFrame.
                     #This operation removes the specified columns from the DataFrame, effectively reducing the number of columns.
df.head()
```

```python
#Renaming Columns
df=df.rename(columns={'MATS3v':'MAT','nHBint10':'NHB'})
df.head()
```

```python
#Dropping the duplicate rows
df.shape #gives (no. of rows , no. of columns) in the dataframe
```

```python
#Finding no. of duplicate rows and columns

#print(df.duplicated) #In pandas, the df.duplicated() function is used to identify duplicate rows in a DataFrame.
                     #It returns a boolean Series that indicates whether each row is a duplicate or not.
                     #The result is a Series of boolean values, where True indicates that the corresponding row is a duplicate and False indicates that it is not.


duplicate_rows=df[df.duplicated()] #df[df.duplicated()] uses the boolean Series returned by df.duplicated() as a filter to select only the rows where the corresponding value is True.
                                  # In other words, it selects the duplicate rows from the DataFrame.

print("(No.of duplicate rows:",duplicate_rows.shape)
```

```python
#Dropping duplicate rows
df=df.drop_duplicates()
df.head()
```

```python
df.count() #used to count the number of non-null values in each column of a DataFrame
```

```python
#NO. the missing or null values

print(df.isnull().sum())    # used to count the number of missing values (null values) in each column of a DataFrame.
                         # It returns a Series where the index represents the column names,
                         # and the values represent the count of missing values in the corresponding columns.
```

```python
#Dropping Misssing or Null values
df=df.dropna() #dropping missing values
df.count()
```

```python
#Detecting Outliers
#An outlier is a point or set of points that are different from other points.
#Sometimes they can be very high or very low.
#It's often a good idea to detect and remove the outliers.
#Because outliers are one of the primary reasons for resulting in a less accurate model. Hence it's a good idea to remove them.
#Often outliers can be seen with visualizations using a box plot.
#Shown below are the box plot of MAT,NHB.
#Herein all the plots, you can find some points are outside the box they are none other than outliers.
```
